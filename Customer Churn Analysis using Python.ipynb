# Customer Churn Distribution

### Count the customers

```python
customer_churn["Churn"].value_counts()
```

```python
customer_churn["Churn"].value_counts().plot(
    kind="bar",
    color=["#4CAF50", "#F44336"])
plt.title("Customer Churn Distribution")
plt.xlabel("Churn")
plt.ylabel("Number of Customers")
plt.savefig("images/churn_distribution.png")
plt.show()
```

## Insight

Most customers remained with the company, while a smaller proportion churned. This indicates that customer retention is generally strong, but reducing churn remains an important business objective.

### Customer Distribution by Contract Type

```python
contract_counts = customer_churn["Contract"].value_counts()
contract_counts.plot(kind="bar", color="skyblue", edgecolor="black")
plt.title("Customer Distribution by Contract Type")
plt.xlabel("Contract Type")
plt.ylabel("Number of Customers")
plt.xticks(rotation=0)
plt.show()
plt.savefig("images/churn_distribution.png")
```

## Insight

Month-to-month contracts account for the largest share of customers. Long-term contracts (one-year and two-year) are less common, suggesting an opportunity to encourage customers to commit to longer plans.

### Customer Distribution by Payment Method

```python
payment = customer_churn["PaymentMethod"].value_counts()
payment.plot(kind="bar", color="orange", edgecolor="black")
plt.title("Customers by Payment Method")
plt.xlabel("Payment Method")
plt.ylabel("Number of Customers")
plt.xticks(rotation=20)
plt.show()
plt.savefig("images/churn_distribution.png")
```

## Insight

Electronic check is the most commonly used payment method, followed by mailed checks and automatic payment options. Understanding customer payment preferences can help improve the payment experience.

### Customer Distribution by Internet Service

```python
internet = customer_churn["InternetService"].value_counts()
internet.plot(kind="pie", autopct="%1.1f%%", startangle=90)
plt.title("Internet Service Distribution")
plt.ylabel("")
plt.show()
plt.savefig("images/churn_distribution.png")
```

## Insight

Fiber optic and DSL are the primary internet services used by customers, while a smaller segment does not subscribe to internet services.

### Distribution of Monthly Charges

```python
plt.hist(customer_churn["MonthlyCharges"], bins=20, color="green", edgecolor="black")
plt.title("Distribution of Monthly Charges")
plt.xlabel("Monthly Charges")
plt.ylabel("Customers")
plt.show()
plt.savefig("images/churn_distribution.png")
```

## Insight

Monthly charges vary across customers, with many concentrated in the lower and middle price ranges. A smaller group pays significantly higher monthly charges.

### Senior Citizen Distribution

```python
senior = customer_churn["SeniorCitizen"].value_counts()
senior.plot(kind="bar", color=["#4CAF50", "#F44336"], edgecolor="black")
plt.title("Senior Citizen Distribution")
plt.xlabel("Senior Citizen (0 = No, 1 = Yes)")
plt.ylabel("Customers")
plt.xticks(rotation=0)
plt.show()
plt.savefig("images/churn_distribution.png")
```

## Insight

Most customers are non-senior citizens, while senior citizens represent a smaller portion of the customer base.

### Customer Gender Distribution

```python
gender = customer_churn["gender"].value_counts()
gender.plot(kind="bar", color=["#42A5F5", "#EC407A"], edgecolor="black")
plt.title("Customer Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Customers")
plt.xticks(rotation=0)
plt.show()
plt.savefig("images/churn_distribution.png")
```

## Insight

The customer base is nearly evenly divided between male and female customers, indicating balanced gender representation.

### Customer Tenure Distribution

```python
plt.hist(customer_churn["tenure"], bins=20, color="purple", edgecolor="black")
plt.title("Customer Tenure Distribution")
plt.xlabel("Tenure (Months)")
plt.ylabel("Customers")
plt.show()
plt.savefig("images/churn_distribution.png")
```
